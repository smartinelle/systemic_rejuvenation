# Exploration

Quick sanity checks of the aging network model. The heavy lifting lives in `src/aging_network` so this notebook just imports and calls high-level functions.

```python
import matplotlib.pyplot as plt
from aging_network import default_simulation_config, plot_mean_X_D_over_time, run_sim

sim_cfg = default_simulation_config()
baseline = run_sim("none", sim_config=sim_cfg)
ax = plot_mean_X_D_over_time(
    baseline,
    func_threshold=sim_cfg.func_threshold,
    death_threshold=sim_cfg.death_threshold,
    title="Baseline trajectory",
)
plt.show()
```

```python
# Compare a couple of interventions
for mode in ["exercise", "drug"]:
    result = run_sim(mode, sim_config=sim_cfg)
    plot_mean_X_D_over_time(
        result,
        func_threshold=sim_cfg.func_threshold,
        death_threshold=sim_cfg.death_threshold,
        title=f"Trajectory: {mode}",
    )
plt.show()
```
